```python
import re
import matplotlib.pyplot as plt

# 打开记录文件
with open('cstest2.txt', 'r', encoding='utf-8') as file:  # 文件名为 channel.sounding.output.txt
    data = file.read()

# 使用正则表达式提取 "Phase-Based Ranging method" 的值
pattern = r"Phase-Based Ranging method: ([\d.]+) meters"
distances = list(map(float, re.findall(pattern, data)))  # 提取后转为浮点数列表

# 打印提取的结果
print("提取的测量结果：")
for i, distance in enumerate(distances, start=1):
    print(f"测量 {i}: {distance} meters")

# 绘制图表

cm=1/2.54  # cm to inch
pt = 1/72.27  # point to inch
linewidth = 246*pt
textwidth = 506*pt  # 0.5 cm in points
axis_label_size = 18   # 坐标轴标签字体
title_size = 20        # 图标题字体
legend_size = 16       # 图例字体
tick_size = 14         # 坐标轴刻度字体
scale_size = 3 #缩放倍数
spine_width = 2 # 边框宽度

plt.figure(figsize=(scale_size*linewidth, scale_size* 0.8 *linewidth), dpi=300)  # 设置图表大小和分辨率
plt.tick_params(axis='both', which='major', labelsize=tick_size)  # 设置坐标轴刻度字体大小
plt.plot(range(1, len(distances) + 1), distances, marker='o', linestyle='-', color='blue', label="Phase-Based Ranging (PBR)")
plt.title("Phase-Based Ranging Measurements", fontsize=title_size)
plt.xlabel("Subevent Counter", fontsize=axis_label_size)
plt.ylabel("Distance (m)", fontsize=axis_label_size)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=legend_size)
plt.tight_layout()

plt.savefig('./poster.fig/phase_based_ranging.svg', dpi=300)  # 保存图表为 PNG 文件

# 显示图表
plt.show()
```

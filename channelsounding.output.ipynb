```python
import re
import matplotlib.pyplot as plt

# 打开记录文件
with open('cstest2.txt', 'r', encoding='utf-8') as file:  # 文件名为 channel.sounding.output.txt
    data = file.read()

# 使用正则表达式提取 "Phase-Based Ranging method" 的值
pattern = r"Phase-Based Ranging method: ([\d.]+) meters"
distances = list(map(float, re.findall(pattern, data)))  # 提取后转为浮点数列表

# 打印提取的结果
print("提取的测量结果：")
for i, distance in enumerate(distances, start=1):
    print(f"测量 {i}: {distance} meters")

# 绘制图表
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(distances) + 1), distances, marker='o', linestyle='-', color='blue', label="Phase-Based Ranging")
plt.title("Phase-Based Ranging Measurements", fontsize=16)
plt.xlabel("Measurement Number", fontsize=12)
plt.ylabel("Distance (meters)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()

# 显示图表
plt.show()
```
